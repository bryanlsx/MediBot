```python
from googletrans import Translator

translator = Translator()
```

```python
import json

with open('intents.json') as file:
  data = json.load(file)

for t in data['intents']:
  for p in t['patterns']:
    r = translator.translate(p, src='en', dest = 'ms')
    print(r.text.lower())
```

```python
for t in data['intents']:
  for p in t['responses']:
    r = translator.translate(p, src='en', dest = 'ms')
    print(r.text.lower())
```
